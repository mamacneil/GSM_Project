# SuperRanker analysis

Performs seqential rank agreement methods of Ekstrøm CT, Gerds TA, Jensen, AK (2018). ``Sequential rank agreement methods for compari-
son of ranked lists.'' _Biostatistics_, *20*(4), 582-598 <doi:10.1093/biostatistics/kxy017> on global apparent consumption rank for shark and ray meat vs overall seafood.

```r
# Import super ranker
library(SuperRanker)
```

```r
# Import ranked consumption data
rdata = read.csv('apparent_consumption.csv')
```

```r
rdata
```

```r
# Sequential rank agreement
SRA = SuperRanker:::sracppfull(as.matrix(rdata[c('elasmos_rank','seafood_rank')]))
```

```r
rdata$SRA = SRA$sra
```

```r
plot(SuperRanker:::sra(as.matrix(rdata[c('elasmos_rank','seafood_rank')])))
```
