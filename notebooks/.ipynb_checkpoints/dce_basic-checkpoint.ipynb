```python
from bisect import bisect
import arviz as az
import numpy as np
import pandas as pd
#from cmdstanpy import CmdStanModel
from matplotlib import pyplot as plt
import pymc as pm
from matplotlib.lines import Line2D
np.set_printoptions(suppress=True)
```

```python
# PYMC implementation of Jim Savages Example
# https://khakieconomics.github.io/2019/03/17/Putting-it-all-together.html

# number of people
N = 100
# number of choices - savage has this be more random for each person 
C = 3
# number of attributes per choice
A = 4
# Create random characteristic matrix (X) in wide format
#np.random.seed(56)

# this matrix represents a series of random attributes for eah choice, in wide format 
# if there are 3 choices each with 4 attributes, the cols 0:3 are for choice 1, 4:7 are for choice 2 etc
#X = np.random.normal(loc=0, scale=1, size=C*A*N).reshape((N,C*A)) # random numbers
X = np.random.choice(np.arange(0,2), size=A*C*N, replace=True).reshape(N, C*A) # dummy variables, jim uses dummy variables

X_df = pd.DataFrame(data=X)
# simulate some data
sim_data = pd.DataFrame.from_dict({
    'customer_id': np.arange(0, N),
    'choice': np.zeros(N),
})

sim_data = pd.merge(sim_data, X_df, left_index=True, right_index=True)
sim_data
```

```python
# simulate betas 3 attributes
#np.random.seed(56)
#betas = np.random.normal(loc=0, scale=1, size=A)
betas = np.array([-2,-.7, 0.4, 2.5])

# calc utility
out_good = np.zeros(N)
u1 = betas[0]*sim_data[0] + betas[1]*sim_data[1] + betas[2]*sim_data[2] + betas[3]*sim_data[3]
u2 = betas[0]*sim_data[4] + betas[1]*sim_data[5] + betas[2]*sim_data[6] + betas[3]*sim_data[7]
u3 = betas[0]*sim_data[8] + betas[1]*sim_data[9] + betas[2]*sim_data[10] + betas[3]*sim_data[11]

# stack
#uts = np.asarray([out_good, u1, u2, u3]).T
uts = np.asarray([u1, u2, u3]).T
# probability
probs = pm.math.softmax(uts, axis=1).eval()
#choice_list = ['Nothing','Prod1', 'Prod2', 'Prod3']
choice_list = ['Prod1', 'Prod2', 'Prod3']
for i in range(0, N):
    choice = np.random.multinomial(n = 1, pvals=probs[i])
    # select string value for scenario based on the index of their choice
    scen_choice = choice_list[np.where(choice == 1)[0][0]]
    # add value to dataframe
    sim_data.loc[i,'choice'] = scen_choice
sim_data
```

```python
probs
```

```python
# the basic model (no intercepts for utility functions, no hierarchy for betas)
N = sim_data.shape[0]
observed = pd.Categorical(sim_data["choice"]).codes
coords = {
    #"alts_probs": ['Nothing','Prod1', 'Prod2', 'Prod3'],
    "alts_probs": ['Prod1', 'Prod2', 'Prod3'],
    "obs": range(N),
}

with pm.Model(coords=coords) as model_1:
    beta_1 = pm.Normal("beta_1", 0, 1)
    beta_2 = pm.Normal("beta_2", 0, 1)
    beta_3 = pm.Normal("beta_3", 0, 1)
    beta_4 = pm.Normal("beta_4", 0, 1)


    ## Construct Utility matrix and Pivot
    u_1 = beta_1*sim_data[0] + beta_2*sim_data[1] + beta_3*sim_data[2] + beta_4*sim_data[3]
    u_2 = beta_1*sim_data[4] + beta_2*sim_data[5] + beta_3*sim_data[6] + beta_4*sim_data[7]
    #u_3 = beta_1*sim_data[8] + beta_2*sim_data[9] + beta_3*sim_data[10] + beta_4*sim_data[11]
    u_none = np.zeros(N)  # Outside Good
    s = pm.math.stack([u_1, u_2, u_none]).T

    ## Apply Softmax Transform
    p_ = pm.Deterministic("p", pm.math.softmax(s, axis=1), dims=("obs", "alts_probs"))

    ## Likelihood
    choice_obs = pm.Categorical("y_cat", p=p_, observed=observed, dims="obs")

    idata_m1 = pm.sample_prior_predictive()
    idata_m1.extend(
        pm.sample(idata_kwargs={"log_likelihood": True}, random_seed=101)
    )
    idata_m1.extend(pm.sample_posterior_predictive(idata_m1))
```

```python
az.summary(idata_m1, var_names=['beta'], filter_vars='regex')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 10))
ax = axs[0]
counts = sim_data.groupby("choice")["customer_id"].count()
predicted_shares = idata_m1["posterior"]["p"].mean(dim=["chain", "draw", "obs"])
ci_lb = idata_m1["posterior"]["p"].quantile(0.025, dim=["chain", "draw", "obs"])
ci_ub = idata_m1["posterior"]["p"].quantile(0.975, dim=["chain", "draw", "obs"])

ax.scatter(ci_lb, ['Nothing','Prod1', 'Prod2', 'Prod3'], color="k", s=2)
ax.scatter(ci_ub, ['Nothing','Prod1', 'Prod2', 'Prod3'], color="k", s=2)
ax.scatter(
    counts / counts.sum(),
    ['Nothing','Prod1', 'Prod2', 'Prod3'],
    label="Observed Shares",
    color="red",
    s=100,
)
ax.hlines(
    ['Nothing','Prod1', 'Prod2', 'Prod3'], ci_lb, ci_ub, label="Predicted 95% Interval", color="black"
)
ax.legend()
ax.set_title("Observed V Predicted Shares")
az.plot_ppc(idata_m1, ax=axs[1])
axs[1].set_title("Posterior Predictive Checks")
ax.set_xlabel("Shares")
ax.set_ylabel("Scenario");
```
