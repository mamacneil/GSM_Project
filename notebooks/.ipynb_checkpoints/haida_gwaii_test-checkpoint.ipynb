```python
from bisect import bisect
import arviz as az
import numpy as np
import pandas as pd
#from cmdstanpy import CmdStanModel
from matplotlib import pyplot as plt
import pymc as pm
from matplotlib.lines import Line2D
#import nest_asyncio
#nest_asyncio.apply()
np.set_printoptions(suppress=True)
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
# Basic Discrete Choice Model
# Based on species density utility values for ONE region (Massett)

# number people
n_customers = 1000
# number of choices i.e. scenarios
P = 3

# create some demographics
# estimate age-range
ages = np.random.lognormal(mean=3.5, sigma=0.4, size=n_customers).round()
bins = np.array([1, 10, 18, 25, 35, 45, 55, 65, 75, 150])

people_df = pd.DataFrame(data=ages, columns=['ages'])
people_df["range"] = pd.cut(people_df.ages, bins)
age_ranges = pd.factorize(people_df.range)[0] + 1

# select a home community
# home communities
communities = [
        'masset',
        'old masset',
        'tow hill',
        'port clements',
        'tlell',
        'miller creek',
        'skidegate',
        'sandspit',
        'daajing giids',
    ]
home_communities = np.random.choice(communities, size=n_customers, p=[0.38, 0.02, 0.03, 0.03, 0.03, 0.03, 0.38, 0.02, 0.08])
people_df['communities'] = home_communities
communities = pd.factorize(people_df.communities)[0] + 1
```

```python
# build up simulated dataframe
sim_data = pd.DataFrame.from_dict({
    'customer_id': np.arange(0,n_customers),
    'age': age_ranges.tolist(),
    'location': communities.tolist(),
    'choice': np.repeat('NA', n_customers),
    'urchins_no': np.repeat(.5, n_customers),
    'urchins_med': np.repeat(1, n_customers),
    'urchins_high': np.repeat(1.5, n_customers),
    'clams_no': np.repeat(0.02, n_customers),
    'clams_med': np.repeat(0.2, n_customers),
    'clams_high': np.repeat(0.7, n_customers),
    'rockfish_no': np.repeat(1.12, n_customers),
    'rockfish_med': np.repeat(0.42, n_customers),
    'rockfish_high': np.repeat(0.08, n_customers),
    'otters_no': np.repeat(0.14, n_customers),
    'otters_med': np.repeat(0.04, n_customers),
    'otters_high': np.repeat(0.01, n_customers),
})
sim_data
```

```python
# estimate utilities and get people to make choice
# generate random betas for 6 attributes 
np.random.seed(53)

B_age = -1.4
B_loc = -0.5
B_urch = 1.4
B_clam = .8
B_rock = .7
B_ott = .4

beta_true = [B_age, B_loc, B_urch, B_clam, B_rock, B_ott]

# calculate utilities
#u_no = beta_true[0]*sim_data['age'] + beta_true[1]*sim_data['location'] + beta_true[2]*sim_data['urchins_no'] + beta_true[3]*sim_data['clams_no'] + beta_true[4]*sim_data['rockfish_no'] + beta_true[5]*sim_data['otters_no']
#u_med = beta_true[0]*sim_data['age'] + beta_true[1]*sim_data['location'] + beta_true[2]*sim_data['urchins_med'] + beta_true[3]*sim_data['clams_med'] + beta_true[4]*sim_data['rockfish_med'] + beta_true[5]*sim_data['otters_med']
#u_high = beta_true[0]*sim_data['age'] + beta_true[1]*sim_data['location'] + beta_true[2]*sim_data['urchins_high'] + beta_true[3]*sim_data['clams_high'] + beta_true[4]*sim_data['rockfish_high'] + beta_true[5]*sim_data['otters_high']

# calculate utilities
u_no = B_urch*sim_data['urchins_no']+B_rock*sim_data['rockfish_no']
u_med =  B_urch*sim_data['urchins_med']+B_rock*sim_data['rockfish_med']
u_high = B_urch*sim_data['urchins_high']+B_rock*sim_data['rockfish_high']

# calculate utilities
#u_no = beta_true[2]*sim_data['urchins_no'] + beta_true[3]*sim_data['clams_no'] + beta_true[4]*sim_data['rockfish_no'] + beta_true[5]*sim_data['otters_no']
#u_med =  beta_true[2]*sim_data['urchins_med'] + beta_true[3]*sim_data['clams_med'] + beta_true[4]*sim_data['rockfish_med'] + beta_true[5]*sim_data['otters_med']
#u_high = beta_true[2]*sim_data['urchins_high'] + beta_true[3]*sim_data['clams_high'] + beta_true[4]*sim_data['rockfish_high'] + beta_true[5]*sim_data['otters_high']



# stack utility functions
utilities = np.asarray([u_no, u_med, u_high]).T
# softmax
soft = pm.math.softmax(utilities, axis=1).eval()

# loop through each customer to make a choice
for i in range(0, n_customers):
    # what scenario do they choose
    choice = np.random.multinomial(n=1, pvals=soft[i])
    # list of scenario string values
    scenario_list = ['no', 'med', 'high']
    # select string value for scenario based on the index of their choice
    scen_choice = scenario_list[np.where(choice == 1)[0][0]]
    # add value to dataframe
    sim_data.loc[i,'choice'] = scen_choice
sim_data
```

```python
choice = np.array([np.random.multinomial(n=1, pvals=s) for s in soft])
```

```python
# Put choice into categorical array
choice = np.where(choice==1)[1]
choice
```

```python
# the basic model (no intercepts for utility functions, no hierarchy for betas)
N = sim_data.shape[0]
observed = pd.Categorical(sim_data["choice"]).codes
coords = {
    "alts_probs": ["no", "med", "high"],
    "obs": range(N),
}

with pm.Model(coords=coords) as model_1:
    intercept = pm.Normal("intercept", 0, 1, shape=2)
    #beta_age = pm.Normal("beta_age", 0, .5)
    #beta_location = pm.Normal("beta_location", 0, .5)
    beta_urchins = pm.Normal("beta_urchins", 0, .5)
    #beta_clams = pm.Normal("beta_clams", 0, .5)
    beta_rockfish = pm.Normal("beta_rockfish", 0, .5)
    #beta_otters = pm.Normal("beta_otters", 0, .5)

    ## Construct Utility matrix and Pivot
    #u_medium = intercept[0] + beta_age*sim_data['age'] + beta_location*sim_data['location'] + beta_urchins*sim_data['urchins_med'] + beta_clams*sim_data['clams_med'] + beta_rockfish*sim_data['rockfish_med'] + beta_otters*sim_data['otters_med']
    #u_high = intercept[1] + beta_age*sim_data['age'] + beta_location*sim_data['location'] + beta_urchins*sim_data['urchins_high'] + beta_clams*sim_data['urchins_high'] + beta_rockfish*sim_data['urchins_high'] + beta_otters*sim_data['urchins_high']
    
    ## Construct Utility matrix and Pivot
    u_no = intercept[0] + beta_urchins*sim_data['urchins_no']+ beta_rockfish*sim_data['rockfish_no'] 
    u_medium = intercept[1] + beta_urchins*sim_data['urchins_med'] + beta_rockfish*sim_data['rockfish_med'] 
    #u_high = intercept[1] + beta_urchins*sim_data['urchins_high']
    
    #u_medium = beta_urchins*sim_data['urchins_med'] + beta_clams*sim_data['clams_med'] + beta_rockfish*sim_data['rockfish_med'] + beta_otters*sim_data['otters_med']
    #u_high = beta_urchins*sim_data['urchins_high'] + beta_clams*sim_data['urchins_high'] + beta_rockfish*sim_data['urchins_high'] + beta_otters*sim_data['urchins_high']
        
    u_high = np.zeros(N)  # Outside Good
    s = pm.math.stack([u_no, u_medium, u_high]).T

    ## Apply Softmax Transform
    p_ = pm.Deterministic("p", pm.math.softmax(s, axis=1))

    ## Likelihood
    choice_obs = pm.Categorical("y_cat", p=p_, observed=choice, dims="obs")

    idata_m1 = pm.sample_prior_predictive()
    idata_m1.extend(
        pm.sample(idata_kwargs={"log_likelihood": True})
    )
    #idata_m1.extend(pm.sample_posterior_predictive(idata_m1))
```

```python
summary = az.summary(idata_m1, var_names=['beta'], filter_vars='regex')
summary
```

```python
# Observed vs Predicted 
fig, axs = plt.subplots(1, 2, figsize=(20, 10))
ax = axs[0]
counts = sim_data.groupby("choice")["customer_id"].count()
predicted_shares = idata_m1["posterior"]["p"].mean(dim=["chain", "draw", "obs"])
ci_lb = idata_m1["posterior"]["p"].quantile(0.025, dim=["chain", "draw", "obs"])
ci_ub = idata_m1["posterior"]["p"].quantile(0.975, dim=["chain", "draw", "obs"])

ax.scatter(ci_lb, ["no", "med", "high"], color="k", s=2)
ax.scatter(ci_ub, ["no", "med", "high"], color="k", s=2)
ax.scatter(
    counts / counts.sum(),
    ["no", "med", "high"],
    label="Observed Shares",
    color="red",
    s=100,
)
ax.hlines(
    ["no", "med", "high"], ci_lb, ci_ub, label="Predicted 95% Interval", color="black"
)
ax.legend()
ax.set_title("Observed V Predicted Shares")
az.plot_ppc(idata_m1, ax=axs[1])
axs[1].set_title("Posterior Predictive Checks")
ax.set_xlabel("Shares")
ax.set_ylabel("Scenario");
```
