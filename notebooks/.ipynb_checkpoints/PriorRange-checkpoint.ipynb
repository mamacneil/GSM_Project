# Explortation of prior log-odds


```python
#!/usr/bin/env python
# coding: utf-8

# # Model for shark and ray meat landings and trade applied to 2014-2019 data

# In[1]:


import os
import pdb

import arviz as az
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pymc as pm
import pytensor.tensor as pyt
import rdata as rd
import seaborn as sns
import xarray as xr
import scipy as sp
from matplotlib.gridspec import GridSpec

# Set figure style.
az.style.use("arviz-darkgrid")
bd = os.getcwd() + "/../Data/"

# Helper functions
def indexall(L):
    poo = []
    for p in L:
        if not p in poo:
            poo.append(p)
    Ix = np.array([poo.index(p) for p in L])
    return poo, Ix


# Helper functions
match = lambda a, b: np.array([b.index(x) if x in b else None for x in a])


def unique(series: pd.Series):
    "Helper function to sort and isolate unique values of a Pandas Series"
    return series.sort_values().unique()
```

Need to find good priors corresponding to:

- impossible
- not retained
- rarely retained
- occasionally retained
- retained
- commonly retained
- large quantities retained

for countries that range between 5 to 150 species retained. We also need to pick prior variances that are not too constraining; trial and error says SD=3 so far.

```python
# Constants
num_countries = 50
num_species = 150
max_catch = 30000

# Initialize a matrix of zeros to represent catches for each country and species
catches_matrix = np.zeros((num_countries, num_species))

# Simulate catches
for country in range(num_countries):
    # Randomly decide how many species each country catches (between 5 and 150)
    num_species_caught = np.random.randint(5, num_species + 1)
    
    # Randomly select which species are caught by this country
    species_caught_indices = np.random.choice(num_species, num_species_caught, replace=False)
    
    # Generate highly left-skewed random catch amounts for the selected species
    hi_num = round(num_species_caught*np.random.uniform(0,0.3))
    hi_catches = np.random.pareto(a=2, size=hi_num) * max_catch / 10
    lo_catches = np.random.pareto(a=50, size=num_species_caught-hi_num) * max_catch / 10
    catches = np.append(hi_catches,lo_catches)
    catches = np.clip(catches, 1, max_catch)  # Ensure values are within range [1, max_catch]
    
    # Assign catches to the matrix
    catches_matrix[country, species_caught_indices] = catches

# Convert the matrix to a Pandas DataFrame for easier visualization
catches_df = pd.DataFrame(catches_matrix, columns=[f'Species_{i+1}' for i in range(num_species)],
                          index=[f'Country_{i+1}' for i in range(num_countries)])
# Get country names
country_ = catches_df.index
species_ = catches_df.columns

# Get total catches per country
TotalCatch = catches_matrix.sum(1)
```

```python
# Calculate the proportions of species in each country
proportions_df = catches_df.div(catches_df.sum(axis=1), axis=0).fillna(0)

# Calculate log-odds for each proportion
# Adding a small constant to avoid log(0)
epsilon = 1e-6
log_odds_df = round(np.log(proportions_df / (1 - proportions_df + epsilon) + epsilon)+10)
log_odds_df_x = log_odds_df

# Scale log-odds to elicitation scale - note odd addition
log_odds_df = round(((log_odds_df+min(log_odds_df.values.flatten()))/max(log_odds_df.values.flatten()))*8)

# Re-scale log odds to magnitude scale
#log_odds_df = log_odds_df+np.log(TotalCatch/max(TotalCatch)+0.001)[:,None]


# Number of species observed per country
nspp_country = (1-(catches_matrix==0)*1).sum(1)

```

```python
plt.hist(np.array(log_odds_df_x).flatten());
```

```python
COORDS = {
    "exporter": country_,
    "species": species_
}
```

```python
with pm.Model(coords=COORDS) as model:
    # Log-odds scale model
    plo_numspp = pm.Normal('plo_numspp')
    reliability_PLO = pyt.exp(plo_numspp*np.log(nspp_country))

    #"""
    # Prior log_odds
    latent_logOdds_landings = pm.Normal("latent_logOdds_landings", np.array(log_odds_df_x), reliability_PLO[:,None])
    
    # Species proportions of total landings
    SppProps = pm.Deterministic(
        "SppProps",
        pm.math.softmax(latent_logOdds_landings,axis=1)
    )

    # National average landings
    Latent_landings = pm.Deterministic(
        "Latent_landings",
        SppProps*TotalCatch[:,None]
    )
    
    # Log landings
    Log_landings = pm.Deterministic(
        "Log_landings",
        pyt.log(Latent_landings+0.0001)
    )

    # National average reported log landings
    EX_Log_landings = pm.Deterministic(
        "EX_Log_landings",
        pyt.log(Latent_landings.sum(1)+0.0001)
    )

    # Errors
    Sigma = pm.Exponential('Sigma',1)

    # Likelihood
    #pm.Normal('Yi', Log_landings, Sigma, observed=np.log(catches_matrix+0.001))
    pm.Normal('Yi', EX_Log_landings, Sigma, observed=np.log(catches_matrix.sum(1)+0.001))
    #"""
```

```python
with model:
    trace = pm.sample()
```

```python
plt.scatter(catches_matrix.flatten(),trace.posterior['Latent_landings'].mean(('chain','draw')).to_numpy().flatten())
x = np.linspace(0, max(catches_matrix.flatten()), 100)
y = x
plt.plot(x, y, linestyle='--', color='red');
```
